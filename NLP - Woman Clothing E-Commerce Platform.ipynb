```python
#This function is designed to preprocess text data by removing stopwords, punctuation, spaces, and digits, and then lemmatizing the remaining tokens to their base form
def clean_text(text):
    
    
    doc = nlp(text)
    clean_text =''
    for word in doc:
        if not word.is_stop and not word.is_punct and not word.is_space:
                clean_text = clean_text + str(word.lemma_) + ' '
    clean_text = clean_text.strip()
    return clean_text
```

```python
#Applying the function for cleaning the Review_Text column in the training set
train_set['Review_Text'] = train_set['Review_Text'].apply(clean_text)
```
